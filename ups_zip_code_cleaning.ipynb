```python
import numpy as np
import pandas as pd
import re
```

```python
df= pd.read_csv('data_clean.csv')
df.head()
```

```python
df['receiver_postal'] = df['receiver_postal'].map(str)

for i, rows in df.iterrows():
    ifor_val = df.iloc[i]['receiver_postal']
    if re.match(r"^\d{9}$", ifor_val):
        zipc = ifor_val[:5]
        plus4 = ifor_val[5:]
        strings = [zipc,plus4]
        ifor_val = '-'.join(strings)
    elif re.match(r"^\d{5}-$", ifor_val):
        zipc = ifor_val[:5]
        ifor_val = zipc
    elif re.match(r"^\d{5}-\d{4}$", ifor_val):
        ifor_val = df.iloc[i]['receiver_postal']  
    elif re.match(r"^\d{5}$", ifor_val):
        ifor_val = df.iloc[i]['receiver_postal']
    else:
        ifor_val = 'invalid'
    df.at[i,'receiver_postal'] = ifor_val
    
df['receiver_postal'] = df['receiver_postal'].str[:5]
```
